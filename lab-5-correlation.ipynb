```python
# imports and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Load dataset
df = pd.read_csv(r'data set\current-power.csv',
                 delimiter=",")
df
```

# Scatter plot

```python

x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.show()
```

# Correlation of coefficient of current and power

```python
df.corr()
```

# Regression line

```python
slope , intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

# scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x, y,
            color='red',
            alpha=0.6,
            label='Data points')
# Line plot
plt.plot(x, line,
         color='#126782',
         label=f'Regression line (r={r_value:.3f})')

plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.legend()
plt.show()
```

# Air Quality

```python
# load dataset
air = pd.read_csv(r"data set\air-quality.csv", 
                  delimiter=",")
air
```

# NOx-Humidity

```python
# regression line
x = air['NOx']
y = air['Relative_Humidity'] 
# regression line
slope , intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

# scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x, y,
            color='red',
            alpha=0.2,
            label='Datapoints')
# Line plot
plt.plot(x, line,
         color='#126782',
         label=f'Regression line (r={r_value:.3f})')

plt.title('NOx vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NOx (ug/cu.m)')
plt.legend()
plt.show()
```

# NO2 Humidity

```python
# regression line
x = air['NO2']
y = air['Relative_Humidity'] 

slope , intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

# scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x, y,
            color='red',
            alpha=0.2,
            label='Datapoints')
# Line plot
plt.plot(x, line,
         color='#126782',
         label=f'Regression line (r={r_value:.3f})')

plt.title('NO2 vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('CO2 (ug/cu.m)')
plt.legend()
plt.show()
```
