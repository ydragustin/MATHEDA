# Descriptive Statistics 

```python
#Jaguar VS Panther
resistance = np.array ([
    [10.6,9.1,9.3,9.8,10.5,10.4,9.5,11,10.4,3,9.8],
    [10.1,11,9.1,20,9.2,10.8,9.9,9.2,9.1,9.1,np.nan]
]).T 
resistance
```

```python
# Creating DataFrame from numpy array
df_resistance = pd.DataFrame(resistance,columns=['Jaguar', 'Panther'])
df_resistance
```

```python
# Summary of dataframe
df_resistance.info()
```

```python
type(df_resistance)
```

```python
# Summary of statistics
df_resistance.describe()
```

```python
# Median
jag_median = df_resistance['Jaguar'].median()
pan_median = df_resistance['Panther'].median()

print (f'Jaguar Median: {jag_median:.2f}ohms')
print (f'Panther Median: {pan_median:.2f}ohms')
```

```python
# Mode
jag_mode = df_resistance['Jaguar'].mode()
pan_mode = df_resistance['Panther'].mode()
```

```python
jag_mode
```

```python
jag_mode.info()
```

```python
print('Jaguar Mode(s):')
for i in jag_mode:
    print(f'{i} ohms')
```

```python
pan_mode
```

```python
pan_mode.info()
```

```python
print ('Panther Mode:')
for i in pan_mode:
    print(f'{i} ohms')  
```

```python
# Jaguar IQR
Q1_jag = df_resistance['Jaguar'].quantile(0.25)
Q3_jag = df_resistance['Jaguar'].quantile(0.75)

print(f'Jaguar IQR = {Q3_jag - Q1_jag:.2f} ohms')
```

```python
# Panther IQR 
Q1_pan = df_resistance['Panther'].quantile(0.25)
Q3_pan = df_resistance['Panther'].quantile(0.75)

print(f'Panther IQR = {Q3_pan - Q1_pan:.2f} ohms')
```

# Bar Chart

```python
jag_mean = df_resistance['Jaguar'].mean()
pan_mean = df_resistance['Panther'].mean()

plt.figure(figsize=(7,4))
bars = plt.bar(['Jaguar','Panther'], [jag_mean, pan_mean], color= '#126782')

plt.title('Average Resistance by Machine')
plt.ylabel('Resistance (ohms)')

# bar labels
plt.bar_label(bars, padding=3)
plt.margins(y=0.1)
plt.show()
```
