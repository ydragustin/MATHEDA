# Iterrating Over a List

```python
fruits = ["apple","orange","grape"]
for i in fruits:
    print(i)
    
 
```

```python
numbers = [1,2,3,4,5,6,6,7,9,10]
for n in numbers:
    if n%2==0:
        print(n)

```

```python
for j in range(2,11,2):
    print(j)
```

# Iteraring Over a Dictionary

```python
for char in "Python":
    print(char)
```

```python

# key-value
employee = {
    'name':'Ada', 
    'position':'analyst',
    'salary': 80000

}
```

```python
employee['position']
```

```python
for key in employee:
    print(key)
```

```python
for value in employee.values():
    print(value)
```

```python
# key-value
employee = {
    'name':['Ada','Jose'],
    'position':['analyst','engineer'],
    'salary': [80000,12000]

}
```

```python
employee['salary']
```
